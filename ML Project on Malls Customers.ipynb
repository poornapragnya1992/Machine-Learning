```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mtp
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mall_customers.csv')
```

```python
x = data.iloc[:,[3,4]].values
```

```python
data.head()
```

```python
from sklearn.cluster import KMeans
wcss_list =[]
for i in range(1,11):
  Kmeans = KMeans(n_clusters=i,init='k-means++', random_state=42)
  Kmeans.fit(x)
  wcss_list.append(Kmeans. inertia_)
mtp.plot(range(1,11), wcss_list)
mtp.title ('The elbow method graph')
mtp.xlabel('Number of clusters(K)')
mtp.ylabel ('wcss_list')
mtp.show()
```

```python
kmeans = KMeans (n_clusters =5, init='k-means++', random_state=42)
y_predict = kmeans.fit_predict(x)
```

```python
mtp.scatter(x[y_predict == 0,0], x[y_predict==0,1], s=100, c='blue', label = 'cluster 1')
mtp.scatter (x[y_predict == 1,0],x[y_predict == 1,1], s=100, c='red', label = 'cluster 2')
mtp.scatter(x[y_predict == 2,0], x[y_predict == 2,1], s=100, c='black', label ='cluster 3')
mtp.scatter (x[y_predict == 3,0], x[y_predict == 3,1], s=100, c ='orange', label ='cluster 4')
mtp.scatter (x[y_predict == 4,0], x[y_predict == 4,1], s=100, c='yellow', label ='cluster 5')
mtp.scatter (kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=300, c='green', label='centroid')
mtp.title ('clusters of customers')
mtp.xlabel ('Annual income(K$)')
mtp.ylabel ('Spending Score(1-100)')
mtp.legend()
mtp.show()
```
